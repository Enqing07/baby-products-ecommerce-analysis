### 2.3 Handling Inconsistent Data Entries

```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

# For each categorical column, identify duplicated categories
for col in categorical_cols:
    print(f"\nColumn: {col}")
    # Create a lowercase version to group similar categories
    lower_case_counts = df[col].astype(str).str.strip().str.lower().value_counts()
    
    # Find groups with more than one distinct capitalization
    duplicates = {}
    for category_lower in lower_case_counts.index:
        original_variants = df[df[col].str.strip().str.lower() == category_lower][col].unique()
        if len(original_variants) > 1:
            duplicates[category_lower] = original_variants
    
    if duplicates:
        print("Categories with capitalization differences:")
        for lower_cat, variants in duplicates.items():
            print(f"  '{lower_cat}': {variants}")
    else:
        print("No capitalization duplicates found.")
```

```python
# Apply title case cleaning to all such columns
for col in categorical_cols:
    df[col] = df[col].astype(str).str.strip().str.title()
```

```python
for col in categorical_cols:
    print(f"{col}:")
    print(df[col].unique(), "\n")
```

# 3.0 Feature Engineering

### 3.1 Extracting Time-Based Features

```python
df['Date'] = pd.to_datetime(df['Date'], utc=False) 

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Month_Name'] = df['Date'].dt.month_name()
df['Month_Year'] = pd.to_datetime(df[['Year', 'Month']].assign(DAY=1))
df['Day'] = df['Date'].dt.day
df['Week'] = df['Date'].dt.isocalendar().week
df['Weekday'] = df['Date'].dt.day_name()
df['Hour'] = df['Date'].dt.hour

date_related_cols = ['Date', 'Year', 'Month', 'Month_Name', 'Month_Year','Day', 'Week', 'Weekday', 'Hour']

other_cols = [col for col in df.columns if col not in date_related_cols]
df = df[date_related_cols + other_cols]
df
```

### Time-based filtering

Data for 2023 is only available up to October. Q4 2023 is excluded to ensure consistency in time-based analysis.

```python
df = df[~((df['Date'].dt.year == 2023) & (df['Date'].dt.quarter == 4))]
```

### 3.2 Discount Rate Calculation

```python
# Compute Discount Rate only for rows with non-zero Discounts
df['Discount Rate'] = 0.0 
non_zero_mask = df['Discounts'] != 0

df.loc[non_zero_mask, 'Discount Rate'] = (
    (df.loc[non_zero_mask, 'Discounts'].abs() / df.loc[non_zero_mask, 'Gross sales'])
    .replace([np.inf, -np.inf], np.nan)
    .round(2)
)

# Move 'Discount Rate' column after 'Discounts'
cols = df.columns.tolist()
cols.remove('Discount Rate')
insert_index = cols.index('Discounts') + 1
cols.insert(insert_index, 'Discount Rate')

df = df[cols]

df[(df['Discounts'] != 0)].head(5)
```
